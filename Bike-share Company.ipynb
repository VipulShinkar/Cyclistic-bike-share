```python
# Coverting datatype of "started_at" & "ended_at" columns to "datetime64" type.
df['started_at'] = df['started_at'].astype('datetime64')
df['ended_at'] = df['ended_at'].astype('datetime64')
```

```python
df.info()
```

```python
# Creating New Column "ride_length"
# In this column, each row contains the difference between "starting time" and "ending time" columns in minutes.
df['ride_length'] = (df['ended_at'] - df['started_at'])/pd.Timedelta(minutes=1)
```

```python
# We can see that many rows in ride_length contained negative values. 
# Such errors happened because the "ending time" is earlier than the "starting time" in their respective rows.
# Number of rows containing Negative Values.
df[df['ride_length'] < 0].count()
```

```python
# Also checking ride_length less than one minute
df[df['ride_length'] < 1].count()
```

```python
# Removing rows containing negative values & ride length less than 1 minute.
# Any trips that were below 60 seconds in length are potentially false starts or users trying to re-dock a bike to ensure it was secure.
df = df[df['ride_length'] >= 1]
```

```python
# Extracting Year, Month, Day, Hour details from start and end time attributes
df['year']= df['started_at'].dt.year
```

```python
df.head()
```

```python
# Extract Year, Month, Day, Hour, and Weekday into individual columns
df['year'] = df['started_at'].dt.year
df['month'] = df['started_at'].dt.strftime('%B')  # Convert month to full name
df['day'] = df['started_at'].dt.day
df['hour'] = df['started_at'].dt.hour
df['weekday'] = df['started_at'].dt.strftime('%A')  # Convert weekday to full name
```

The format codes '%B' and '%A' are used in the strftime() method to represent the month and weekday names, respectively.

'%B': This format code represents the full name of the month. For example, 'January', 'February', 'March', etc. It provides the full textual representation of the month.

'%A': This format code represents the full name of the weekday. For example, 'Monday', 'Tuesday', 'Wednesday', etc. It provides the full textual representation of the weekday.

By using these format codes, the extracted month and weekday values from the timestamps are converted into their corresponding full names in string format. However, please note that the availability of format codes may vary across programming languages and libraries. These specific codes (%B and %A) are commonly used in Python's strftime() method to format datetime values.

```python
# Define a function to map month to season
def season(month):
    if month in ['March','April','May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    elif month in ['September', 'October', 'November']:
        return 'Fall'
    else:
        return 'Winter'

# Creating a new 'season' column based on the 'month' column
df['season'] = df['month'].apply(season)
```

```python
# Define a function to map hour to time of day
def time_of_day(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 22:
        return 'Evening'
    else:
        return 'Night'

# Creating a new 'time_of_day' column based on the 'hour' column
df['time_of_day'] = df['hour'].apply(time_of_day)
```
